```python
import seaborn as sns
import matplotlib.pyplot as plt
subjects = ["Math", "Physics", "Chemistry", "Computer", "English"]
marks = [78, 65, 72, 85, 80]
sns.set_style("whitegrid")
sns.barplot(x=subjects, y=marks, palette="Set1", hue=subjects, legend=False)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Subject Wise Performance")
plt.show()
```
